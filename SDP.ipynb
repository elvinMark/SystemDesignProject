```
import pandas as pd 
df = pd.read_csv("2000tags.csv")
tags = [elem[1:] for elem in df["タグ"]]
```

```
tags_vec = [model.get_word_vector(elem) for elem in tags]
df["vector"] = tags_vec
df.to_csv("2000tags_vectors.csv")
```

```
import numpy as np 
data = np.array(tags_vec)
np.savetxt("tags_vector.dat",tags_vec)
```

```
vecs = np.loadtxt("tags_vector.dat")
print(vecs.shape)
```

```
from sklearn import cluster 
k_means = cluster.KMeans(n_clusters=100)
k_means.fit(vecs)
```

```
groups = [[] for i in range(300)]
for i,j in zip(tags,list(k_means.labels_)):
  groups[j].append(i)
```

```
print(groups)
```
