```python
import pandas as pd
import numpy as np
```

```python
# Loading sample data and tags
df1 = pd.read_csv("./data/sample_doc_vec.csv")
df2 = pd.read_csv("./data/2000tags.csv")
```

```python
# Testing on how to get the index of specific tags
tags = df2["タグ"]
x = list(np.where(tags.str.match("#日記$|#エッセイ$|#小説$"))[0])
x
```

```python
# Joining the tags into an string query (as in the prev line)
"|".join([s + "$" for s in eval(df1["tags"][1])])
```

```python
# Convert all numerica tags into numerical ones
df1["labels"] = [list(np.where(tags.str.match("|".join([s + '$' for s in eval(elem)])))[0]) for elem in df1["tags"]]
```
